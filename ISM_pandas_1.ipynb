## **Výběr konkrétních řádků**

#### & - and
#### | - or
#### ~ - not

```python
# Filtrování řádků s vyšší prodejní cenou než je 175000
df_train[df_train.SalePrice > 175000]
```

```python
sum(df_train.SalePrice > 175000) #počet vyfiltrovaných řádků 
```

```python
df_train[(df_train.SalePrice > 175000) & (df_train.SalePrice < 223500)] # každá podmínka musí být zvlášť v závorkách
```

```python
# Výběr řádků dle hodnot v seznamu
df_train[df_train.SalePrice.isin([175000, 250000])]
```

```python
# 
df_train[~((df_train.SalePrice > 175000) & (df_train.SalePrice < 223500))]
```
